```python
import numpy as np
import matplotlib.pyplot as plt

class Linear_Regression:
    def __init__(self,data):
        self.x = data['X']
        # print(self.x)
        self.y = data['y']
        self.dim = np.size(self.x[0])
        self.size = len(self.x)
        self.w = np.ones(self.dim + 1)
        self.lr = 0.001

        self.x = self.sep_x()
        
        for _ in range(1000):
            self.update()

        x = np.reshape(np.array(range(20)),(1,20))
        
        fig = plt.figure(figsize=(10,10))
        plt.scatter(self.x[0], self.y)
        plt.plot(range(20),self.h(x))
        fig.show()
        # print(self.w)
        # print(self.size)
        # print(self.dim)
        # print(self.x)
        # print(rst)

    def sep_x(self):
        if(self.dim == 1):return np.reshape(self.x,(1,self.size))
        rst = np.zeros((self.dim,self.size))
        for i in range(self.size):
            for j in range(self.dim):
                rst[j,i] = self.x[i,j]
        return rst

    def update(self):
        y_h = self.y - self.h(self.x)
        self.w[0] += self.lr * np.sum(y_h)
        for i in range(1,self.dim + 1):
            self.w[i] += self.lr * np.sum(y_h * self.x[i - 1])
        

    def h(self,x):
        rst = self.w[0]
        for i in range(1,self.dim + 1):
            rst += self.w[i] * x[i - 1]
        return rst


data = np.load('data1.npz')
regressor = Linear_Regression(data)
# x = data['X']
# y = data['y']
# plt.scatter(x,y)
# plt.show()
```
